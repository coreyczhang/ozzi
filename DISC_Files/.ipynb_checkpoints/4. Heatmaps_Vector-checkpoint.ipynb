```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
import matplotlib.colors as mcolors
from tqdm import tqdm

# Define your base directory for outputs
png_dir = r'/Users/coreyzhang/Documents/DISC_Files/Output/Cropped'  # PNG files directory
csv_dir = r'/Users/coreyzhang/Documents/DISC_Files/Output/Cropped/csv'  # CSV files directory
output_dir_path = os.path.join(png_dir, 'O', 'OVERLAY')  # Update this path as needed
os.makedirs(output_dir_path, exist_ok=True)

# Settings
OverLay = True
Contour = True
Vector = False
FixedScaleValue = [0, 75]

# Define and create subdirectories for different types of outputs
if Contour:
    heatmap_dir = os.path.join(output_dir_path, 'Heatmap')
    os.makedirs(heatmap_dir, exist_ok=True)
if Vector:
    vector_dir = os.path.join(output_dir_path, 'Vector_folder')
    os.makedirs(vector_dir, exist_ok=True)

# List all CSV and PNG files
csv_files = sorted([f for f in os.listdir(csv_dir) if f.endswith('.csv')])
png_files = sorted([f for f in os.listdir(png_dir) if f.endswith('.png')], reverse = True)

# Ensure the number of CSV and PNG files match
#assert len(csv_files) == len(png_files), "Mismatch in number of CSV and PNG files"

for csv_file, png_file in tqdm(zip(csv_files, png_files), total=len(csv_files), desc="Processing files"):
    print(f"Processing {csv_file} and {png_file}")
    
    csv_file_path = os.path.join(csv_dir, csv_file)
    image_path = os.path.join(png_dir, png_file)
    
    df = pd.read_csv(csv_file_path)
    image = Image.open(image_path)
    empty_image = np.zeros((*image.size[::-1], 4), dtype=np.uint8)

    if Contour:
        FixedScaleValue = [0, 90]
        scale = [i / 100 for i in range(0, int(100 * (FixedScaleValue[1] - FixedScaleValue[0])) + 1, int(100 * (FixedScaleValue[1] - FixedScaleValue[0]) / 25))]
        norm = mcolors.Normalize(vmin=FixedScaleValue[0], vmax=FixedScaleValue[1])

        dpi = image.info['dpi'] if 'dpi' in image.info else (300, 300)
        dpi = (int(dpi[0]), int(dpi[1]))

        # Reshape the data into a 2D grid
        x_unique = np.sort(df['pos_x'].unique())
        y_unique = np.sort(df['pos_y'].unique())
        Z = df.pivot_table(index='pos_y', columns='pos_x', values='disp_xy_indi', fill_value=np.nan).reindex(index=y_unique, columns=x_unique, fill_value=np.nan).values

        fig, ax = plt.subplots(figsize=(image.width / dpi[0], image.height / dpi[1]), dpi=dpi[0])

        if OverLay:
            ax.imshow(image, extent=[x_unique.min(), x_unique.max(), y_unique.max(), y_unique.min()])
        else:
            ax.imshow(empty_image, extent=[x_unique.min(), x_unique.max(), y_unique.max(), y_unique.min()])

        # Mask the array where Z values are NaN
        Z_masked = np.ma.masked_where(np.isnan(Z), Z)

        contour = ax.contourf(x_unique, y_unique, Z_masked, levels=scale, cmap=plt.cm.jet, alpha=0.5, norm=norm, extent=[x_unique.min(), x_unique.max(), y_unique.max(), y_unique.min()])
        ax.axis('off')

        # Superimpose the heatmap onto the original image and save
        combined_output_path = os.path.join(heatmap_dir, os.path.splitext(png_file)[0] + '_heatmap_superimposed.png')
        plt.savefig(combined_output_path, dpi=dpi[0], bbox_inches='tight', pad_inches=0, transparent=True)
        plt.close()

    if Vector:
        plt.figure(figsize=(image.width / 100, image.height / 100))
        if OverLay:
            plt.imshow(image, cmap='gray', extent=[0, image.width, 0, image.height])
        else:
            plt.imshow(empty_image, extent=[0, image.width, 0, image.height])
        plt.quiver(df['pos_x'], df['pos_y'], df['disp_x'], df['disp_y'], angles='xy', scale_units='xy', scale=1, color='red', width=0.0005)
        plt.gca().invert_yaxis()
        plt.axis('equal')
        plt.axis('off')
        vector_output_path = os.path.join(vector_dir, os.path.splitext(png_file)[0] + '_vector.png')
        plt.savefig(vector_output_path, transparent=False, format='png', dpi=1000)
        plt.close()

print("Processing complete.")
```
