```python
import cv2
import os
from tqdm import tqdm

# Set directories and file paths
Dir = '/Users/eugene/Documents'
inDir = os.path.join(Dir, "Mice_Input")
MovFile = 'FJ1_Wounded_01.MOV'
input_path = os.path.join(inDir, MovFile)
outDir = os.path.join(Dir, 'Mice_Output')

# Ensure output directory exists
if not os.path.exists(outDir):
    os.makedirs(outDir)

# Open the video file
vc = cv2.VideoCapture(input_path)
if not vc.isOpened():
    raise Exception(f"Error opening video file: {input_path}")

# Get total frame count for the progress bar
total_frames = int(vc.get(cv2.CAP_PROP_FRAME_COUNT))

# Initialize variables
c = 0  # Frame counter
timeF = 1  # Number of actual frames between two captured frames

# Read and process frames with a progress bar
with tqdm(total=total_frames, desc="Extracting frames", unit="frame") as pbar:
    while True:
        rval, frame = vc.read()
        if not rval:
            break

        if c % timeF == 0:
            filename = os.path.join(outDir, f'{1000000 + c:07d}.png')
            cv2.imwrite(filename, frame)
        
        c += 1
        pbar.update(1)
        cv2.waitKey(1)

# Release the video capture object
vc.release()

print(f"Total frames captured: {c}")
```
