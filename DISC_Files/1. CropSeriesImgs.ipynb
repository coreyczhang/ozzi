```python
import cv2
import os
from tqdm import tqdm

window_width = 600
window_height = 1200

def draw_box(image):
    # Display the image and allow the user to draw a box
    clone = image.copy()
    cv2.namedWindow("Draw Box", cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Draw Box", window_width, window_height)

    # Initialize variables for the box coordinates
    box_start = None
    box_end = None
    drawing = False
    last_update_time = 0
    update_interval = 0.02  # Update every 20 milliseconds

    def draw_rectangle(event, x, y, flags, param):
        nonlocal box_start, box_end, drawing, clone, last_update_time

        # Handle mouse events
        if event == cv2.EVENT_LBUTTONDOWN:
            box_start = (x, y)
            drawing = True

        elif event == cv2.EVENT_MOUSEMOVE:
            if drawing:
                current_time = time.time()
                if current_time - last_update_time > update_interval:
                    last_update_time = current_time
                    clone = image.copy()
                    cv2.rectangle(clone, box_start, (x, y), (0, 255, 0), 2)
                    cv2.imshow("Draw Box", clone)

        elif event == cv2.EVENT_LBUTTONUP:
            box_end = (x, y)
            drawing = False
            cv2.rectangle(clone, box_start, box_end, (0, 255, 0), 2)
            cv2.imshow("Draw Box", clone)

    # Register the callback function for mouse events
    cv2.setMouseCallback("Draw Box", draw_rectangle)

    cv2.imshow("Draw Box", clone)

    # Wait for the user to draw the box and close the window
    while True:
        key = cv2.waitKey(1) & 0xFF

        if key == ord("r"):
            # Reset the box if the 'r' key is pressed
            clone = image.copy()
            cv2.imshow("Draw Box", clone)

        elif key == ord("c"):
            # Close the window and proceed to crop the images if the 'c' key is pressed
            break

    cv2.destroyAllWindows()

    # Return the box coordinates
    return box_start, box_end

def crop_and_save_images(inDir, box_start, box_end, image_files):
    # Crop each image and save it
    outDir = os.path.join(inDir, 'Cropped')
    os.makedirs(outDir, exist_ok=True)
    print("Cropping process initiated...")
    for i, image_file in enumerate(tqdm(image_files, desc="Processing images", unit="file")):
        image_path = os.path.join(inDir, image_file)
        image = cv2.imread(image_path)
        if image is None:
            print(f"Error loading image: {image_path}")
            continue
        cropped_image = image[box_start[1]:box_end[1], box_start[0]:box_end[0]]
        cv2.imwrite(os.path.join(outDir, f"cropped_{image_file}"), cropped_image)
    print(f"Total frames processed: {len(image_files)}")

# Directory containing the images
inDir = r"/Users/coreyzhang/Documents/DISC_Files/Output"  # Replace with the actual directory path

# Get a list of image file names in the directory
image_files = [f for f in os.listdir(inDir) if f.endswith('.png')]

# Ensure there are images to process
if not image_files:
    raise Exception(f"No PNG images found in directory: {inDir}")

# Load only the first image
first_image = cv2.imread(os.path.join(inDir, image_files[0]))
if first_image is None:
    raise Exception(f"Error loading the first image: {os.path.join(inDir, image_files[0])}")

# Draw a box on the first image
box_start, box_end = draw_box(first_image)

# Crop and save the series of images
crop_and_save_images(inDir, box_start, box_end, image_files)

print("Cropping process completed.")
```
